#### 1. В чём различие между зависимыми и независимыми выборками?

```python
Зависимые выборки в отличие от независимых имеют гоморфную пару, т.е. одному событию из выборки X 
соответствует одно и только одно событие в выборке Y и наоборот. 
Зависимые выборки должны иметь одинаковый объём в отличие от независимых. 
Пример зависимых выборок: выборка значений температуры больного до приёма лекарств и выборка после приёма лекартв,
мужья и жёны.
Пример независимых выборок: кошки и собаки, мужчины и женщины...
```

#### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

```python
Параметрические критерии применяются на выборках которые имеют нормальное распределение,
а непараметрические когда выборка не имеет нормального распределения или когда его достаточно сложно рассчитать.
```
